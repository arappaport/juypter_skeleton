```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# initialise data of lists.
data = {'Name':[ 'Mohe' , 'Karnal' , 'Yrik' , 'jack' ],
        'Age':[ 30 , 21 , 29 , 28 ],
        'Weight':[ 130 , 121 , 129 , 200 ]}

# Create DataFrame
df = pd.DataFrame( data )

# Print the output.
df
```

```python
#Examples from:
#https://www.geeksforgeeks.org/data-visualization-with-python-seaborn/#Line%20Plot

# Selecting style as white,
# dark, whitegrid, darkgrid
# or ticks
sns.set( style = "white" )

# Generate a random uni-variate dataset
rs = np.random.RandomState( 10 )
d = rs.normal( size = 50 )
```

```python
# Plot a simple histogram and kde
# with binsize determined automatically
sns.distplot(d, kde = True, color = "g")
```

```python
# Load the example mpg dataset
mpg = sns.load_dataset("mpg")

# Plot miles per gallon against horsepower with other semantics
sns.relplot(x="horsepower", y="mpg", hue="origin", size="weight",
            sizes=(400, 40), alpha=.5, palette="muted",
            height=6, data=mpg)
```
